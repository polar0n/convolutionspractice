```python
from PIL import Image
from matplotlib import pyplot as plt
import numpy as np
from scipy.signal import fftconvolve
```

```python
img = Image.open('images/frame.jpg').convert('L')
img.save('images/frame_gray.png')
```

```python
img = np.array(Image.open('images/frame_gray.png'))
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
```

```python
X = 1/5
H = -1/16
kernel = np.array([
    [0, H, H, H, 0],
    [H, H, X, H, H],
    [H, X, X, X, H],
    [H, H, X, H, H],
    [0, H, H, H, 0]
])
```

```python
conv_image = fftconvolve(img, kernel)
```

```python
new_image = np.zeros(
    tuple(conv_image.shape) + (3,),
    dtype=np.uint8
)

red = conv_image.copy()
np.place(red, red > 0, 0)
red = abs(red)
np.place(red, red > np.ceil(red.flatten().mean()), 255)
np.place(red, red <= np.ceil(red.flatten().mean()), 0)

blue = conv_image.copy()
np.place(blue, blue < 0, 0)
np.place(blue, blue > np.ceil(blue.flatten().mean()), 255)
np.place(blue, blue <= np.ceil(blue.flatten().mean()), 0)

new_image[:,:,0] = red.astype(np.uint8)
new_image[:,:,2] = blue.astype(np.uint8)
```

```python
plt.imshow(new_image)
```

```python
Image.fromarray(new_image).save('images/lines.png')
```

```python
new_image = np.zeros(
    tuple(conv_image.shape) + (3,),
    dtype=np.uint8
)

pixels = conv_image.astype(np.int8).copy()
np.place(pixels, pixels > 0, 0)
pixels = abs(pixels).astype(np.uint8)
np.place(pixels, pixels > np.ceil(pixels.flatten().mean()), 255)
np.place(pixels, pixels <= np.ceil(pixels.flatten().mean()), 0)

new_image[:,:,0] = pixels
new_image[:,:,1] = pixels
new_image[:,:,2] = pixels
```

```python
plt.imshow(pixels, cmap='gray', vmin=0, vmax=255)
```

```python
Image.fromarray(new_image).save('images/lines_white.png')
```
